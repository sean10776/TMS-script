```python
import psutil
from scapy.all import sniff
from scapy.packet import Packet
from scapy.plist import PacketList
```

```python

def packet_handler(packet:Packet):
    # Process the packet data here
    print(packet.sprintf("{IP:%IP.src% -> %IP.dst%\n}{Raw:%Raw.load%\n}"))

def sniff_process_traffic(process_name):
    # Find the process ID of the target process
    process_pid = None
    for proc in psutil.process_iter(['pid', 'name']):
        if process_name.lower() in proc.info['name'].lower():
            process_pid = proc.info['pid']
            break

    if process_pid is None:
        print(f"Process '{process_name}' not found.")
        return

    print(f"Sniffing traffic for process '{process_name}' (PID: {process_pid})...")

    try:
        # Sniff traffic and apply the packet handler function to each packet
        sniff(filter=f"tcp and host 202.80.104.142", prn=packet_handler, count=0)
    except KeyboardInterrupt:
        print("Sniffing stopped.")
sniff_process_traffic("maplestory")
```
